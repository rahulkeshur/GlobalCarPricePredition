```python
X = GlobalCar_DataSet.drop("Price_USD", axis=1)
Y = GlobalCar_DataSet["Price_USD"]
```

```python
## Drop the columns for better output
X = X.drop("Age_Category",axis=1)
X = X.drop("Car_ID",axis=1)
```

```python
X.head()
```

```python
## Label encoding

# This method is not proper, so I changed it
from sklearn.preprocessing import LabelEncoder

##
# le = LabelEncoder()
# X["Brand"] = le.fit_transform(X["Brand"])
# X["Body_Type"] = le.fit_transform(X["Body_Type"])
# X["Fuel_Type"] = le.fit_transform(X["Fuel_Type"])
# X["Transmission"] = le.fit_transform(X["Transmission"])
# X["Manufacturing_Country"] = le.fit_transform(X["Manufacturing_Country"])
# X["Price_Category"] = le.fit_transform(X["Price_Category"])
# X["Age_Category"] = le.fit_transform(X["Age_Category"])
##

le_brand = LabelEncoder()
le_body = LabelEncoder()
le_fuel = LabelEncoder()
le_trans = LabelEncoder()
le_country = LabelEncoder()
le_price = LabelEncoder()

X["Brand"] = le_brand.fit_transform(X["Brand"])
X["Body_Type"] = le_body.fit_transform(X["Body_Type"])
X["Fuel_Type"] = le_fuel.fit_transform(X["Fuel_Type"])
X["Transmission"] = le_trans.fit_transform(X["Transmission"])
X["Manufacturing_Country"] = le_country.fit_transform(X["Manufacturing_Country"])
X["Price_Category"] = le_price.fit_transform(X["Price_Category"])
```

```python
X.head()
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=41)
```

## Import the model for train and test

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rendomForest = RandomForestRegressor()
```

```python
rendomForest.fit(X_train,y_train)
```

```python
rendom_pred = rendomForest.predict(X_test)
```

```python
rendom_pred
```

```python
plt.scatter(y_test,rendom_pred)
```

```python
## residuals
residuals = y_test-rendom_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind = "kde")
```

## Assumptions

```python
## scatter plot with respect to prediction and residuals
plt.scatter(rendom_pred,residuals)
```

```python
## calculation that helps to better vision of how the model trains

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print(mean_absolute_error(y_test,rendom_pred))
print(mean_squared_error(y_test,rendom_pred))
print(np.sqrt(mean_squared_error(y_test, rendom_pred)))
```

## R square and adjusted R square

```python
# Formula 
# R^2 = 1 - SSR/SST
# R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of square
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,rendom_pred)
print(score)
```

```python
# Adjusted R2 = 1 - [(1-R2)*(n-1)/(n-k-1)]

1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## New Data Prediction

```python
# a method that helps to convert a string to an int

def prepare_input(input_data, X, encoders):
    result = []

    for col in X.columns:
        val = input_data[col]

        if col in encoders:  # categorical
            val = encoders[col].transform([val])[0]

        result.append(val)

    return result
```

```python
X.head()
```

```python
encoders = {
    "Brand": le_brand,
    "Body_Type": le_body,
    "Fuel_Type": le_fuel,
    "Transmission": le_trans,
    "Manufacturing_Country": le_country,
    "Price_Category": le_price
    
}

user_input = {
    "Brand": "Mercedes",
    "Manufacture_Year": 2006,
    "Body_Type": "SUV",
    "Fuel_Type": "Petrol",
    "Transmission": "Manual",
    "Engine_CC":4089,
    "Horsepower": 547,
    "Mileage_km_per_l":17,
    "Manufacturing_Country": "USA",
    "Car_Age":20,
    "Price_Category": "Premium",
    "HP_per_CC": 0.1338,
    "Efficiency_Score": 0.35
}

data = prepare_input(user_input, X, encoders)
data = np.array(data).reshape(1,-1)
data
```
